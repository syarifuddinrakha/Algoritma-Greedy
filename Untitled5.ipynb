```python
harga = int(input("Masukkan Total Harga : "))
bayar = int(input("Masukkan Uang Yang Diberikan : "))

if bayar < harga :
  print("Transaksi Gagal : Dana Tidak Mencukupi!")
else:
  kembalian = bayar - harga
  print("Total Kembalian : Rp", kembalian)

  lembar = kembalian // 50000
  kembalian = kembalian % 50000
  if lembar > 0: print("-Rp 50.000 :", lembar, "lembar")

  lembar = kembalian // 20000
  kembalian = kembalian % 20000
  if lembar > 0: print("-Rp 20.000 :", lembar, "lembar")

  lembar = kembalian // 10000
  kembalian = kembalian % 10000
  if lembar > 0: print("-Rp 10.000", lembar , "lembar")

  lembar = kembalian // 5000
  kembalian = kembalian % 5000
  if lembar > 0: print("-Rp 5.000 :", lembar, "lembar")

```
