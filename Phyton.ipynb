```python
print("   SELAMAT DATANG DI WARUNG PINTAR   ")

keranjang_nama = []
keranjang_harga = []
total_belanja = 0

jumlah_barang = int(input("Berapa barang yang akan dibeli? "))
for i in range(jumlah_barang):
  print(f"\n--- masukkan data barang ke-{i+1} ---")
  nama = input("Nama barang: ")
  harga = int(input("Harga barang: Rp "))
  keranjang_nama.append(nama)
  keranjang_harga.append(harga)
  total_belanja = total_belanja + harga

print("\n=== menghitung total ===")
if total_belanja > 50000:
  print("anda mendapatkan diskon 10%")
  diskon = total_belanja * 0.1
  total_bayar = total_belanja - diskon
else:
  print("belanja kurang dari Rp 50.000, tidak dapat diskon")
  total_bayar = total_belanja
print("--------------------------------")
print("Total bersih: Rp", int(total_bayar))

print("\n=== PEMBAYARAN ===")
uang_pelanggan = int(input("Uang yang diberikan: Rp "))
kembalian = uang_pelanggan - total_bayar
print("Kembalian anda: Rp", int(kembalian))
print("teria kasih telah berbelanja!")
print("================================")
```
